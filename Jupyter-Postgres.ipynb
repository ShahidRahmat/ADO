```python
# Cleansing the data

df2 = df.dropna(subset = ['address1','zip','city', 'latitude','longitude','Primary NAICS','amputation'] )
print(df2.isnull().sum())
print(df2.shape)

df2.fillna('missing', inplace=True)
print(df2.isnull().sum())
```

```python
# Adding new columns

others = ['AMERICAN SAMOA', 'DISTRICT OF COLUMBIA', 'GUAM', 'PUERTO RICO', 'VIRGIN ISLANDS', 'NORTHERN MARIANA ISLANDS']
northeast = ['CONNECTICUT', 'MAINE', 'MASSACHUSETTS', 'NEW HAMPSHIRE', 'RHODE ISLAND', 'VERMONT', 'NEW JERSEY', 'NEW YORK', 'PENNSYLVANIA']
midwest = ['ILLINOIS', 'INDIANA', 'MICHIGAN', 'OHIO', 'WISCONSIN', 'IOWA', 'KANSAS', 'MINNESOTA', 'MISSOURI', 'NEBRASKA', 'NORTH DAKOTA', 'SOUTH DAKOTA']
south = ['DELAWARE', 'FLORIDA', 'GEORGIA','MARYLAND','NORTH CAROLINA','SOUTH CAROLINA','VIRGINIA','WEST VIRGINIA','ALABAMA','KENTUCKY','MISSISSIPPI','TENNESSEE','ARKANSAS','LOUISIANA','OKLAHOMA','TEXAS']
west = ['ARIZONA','COLORADO','IDAHO','MONTANA','NEVADA','NEW MEXICO','UTAH','WYOMING','ALASKA','CALIFORNIA','HAWAII','OREGON','WASHINGTON']
```

```python
df2['Region'] = np.nan
for i in df2.index:
    s = df2['State'][i]
    reg = ''
    if s in others:
        reg = 'Others'
    elif s in northeast:
        reg = 'Northeast'
    elif s in midwest:
        reg = 'Midwest'
    elif s in south:
        reg = 'South'
    elif s in west:
        reg = 'West'
    else:
        reg = 'Missing'
    df2['Region'][i] = reg

display(df2.head())
```
